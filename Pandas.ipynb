```python
import numpy as np
df = pd.DataFrame({'Cricket':[1,2,np.nan,4,6,7,2,np.nan],
                   'BaseBall':[5,np.nan,np.nan,5,7,2,4,5],
                   'Tennis':[1,2,3,4,5,6,7,8]})
df
# NaN - Not a Number/null value/missing data
# When ever we get NaN,int gets converted to float
```

```python
#drop all the null values
df.dropna()
```

```python
#replace all the NaN values with zero
df = df.fillna(value = 0)
df
```
