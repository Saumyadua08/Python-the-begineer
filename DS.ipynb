```python
# list
li=['a','b','c']
li
```

```python
# adding a new element
li.append('d')
li
```

```python
# adding a new element at any index
li.insert(2,'new')
li
```

```python
# removing an element based on index
li.pop(2)
li
```

```python
# removing an element with name
li.remove('a')
li
```

```python
# reversing a list
li.reverse()
li
```

```python
# creating a new list
li2= [1,2,3]
li2
```

```python
# concatenating two lists
li+li2  
```

```python
# Another way to combine li with li2
li.extend(li2)
li
```

```python
li*2
```

```python
# length of the list
len(li)
```

```python
# Tuple
tup= (1,2,3,4,5)
tup
```

```python
# Tuple is immutable
tup.append(6)
```

```python
# Set
set1= {1,2,3,4,5}
set1
```

```python
set2= {6,7,8}
```

```python
# union of sets
set1 | set2   
```

```python
# We can also do union alternatively,
a=set1.union(set2)
print(a)
```

```python
# or we can simply update it
set1.update(set2)
set1
```

```python
# intersection of sets
set1 & set2
```
